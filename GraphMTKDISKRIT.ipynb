```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, WeekdayLocator
import datetime 


# Mengimput data dari website owid

df = pd.read_csv(
    'https://covid.ourworldindata.org/data/owid-covid-data.csv', 
    usecols=['date', 'location', 'total_deaths'], 
    parse_dates=['date']) 



countries = ['Indonesia', 'Ireland', 'Japan', 'Canada'] # Negara yang di pilih
df = df[df['location'].isin(countries)]


# Membuat tabel pivot yang akan di konversi kedalam grafik

pivot = pd.pivot_table(
    data=df,                                    
    index='date',                               
    columns='location',                         
    values='total_deaths',    
    aggfunc='mean',                             
    )

pivot = pivot.fillna(method='ffill')





# Memasukan negara-negara ke dalam grafik

fig, ax = plt.subplots(figsize=(12,5))


for country in countries:
    ax.plot(
        pivot.index,              # Apa yang di tulis di row X (kosong)
        pivot[country],           # Apa yang di tulis di row Y (Negara)
        
    )
    ax.text(
        x = pivot.index[-1] + timedelta(days=2),    
        y = pivot[country].max(),                   
        s = country,                                
        
    )

date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_locator(WeekdayLocator(byweekday=(0), interval=8))
ax.xaxis.set_major_formatter(date_form)
ax.set_xlim([datetime.date(2020, 2, 24), datetime.date(2021, 9, 19)]) # Aturan jarak tanggal


plt.show()
```
